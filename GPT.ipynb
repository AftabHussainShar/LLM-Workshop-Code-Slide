```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load pre-trained model and tokenizer
model_name = "gpt2"  # Specify the pre-trained model to use
tokenizer = GPT2Tokenizer.from_pretrained(model_name)  # Load tokenizer
model = GPT2LMHeadModel.from_pretrained(model_name)    # Load model

# Define input text
input_text = "i love to eat pizza"  # Input prompt for text generation

# Tokenize input text
input_ids = tokenizer.encode(input_text, return_tensors="pt")  # Convert input text to token IDs

# Generate text
output = model.generate(input_ids,              # Input token IDs
                         max_length=50,         # Maximum length of generated text
                         num_return_sequences=1,# Number of text sequences to generate
                         temperature=0.6,       # Controls randomness of generation (higher values = more randomness)
                         do_sample=True,  # Enable sampling-based generation
                         pad_token_id=tokenizer.eos_token_id)  # ID of the padding token

# Decode and print generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)  # Convert token IDs back to text
print("Generated Text:", generated_text)  # Print the generated text
```
